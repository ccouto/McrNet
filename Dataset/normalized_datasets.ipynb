```python
import pandas as pd
import numpy as np
```

```python
def Mcr_u(hw1,hw2,tw,bf1,tf1, bf2, tf2,L):
    #input dimensions in mm
    r=0
    h=(hw1+hw2)/2+tf1+tf2
    b=(bf1+bf2)/2
    tf=(tf1+tf2)/2

    Iz = 1 / 12 * (2 * tf * b ** 3 + (h - 2 * tf) * tw ** 3) + 0.03 * r ** 4 + 0.2146 * r ** 2 * (tw + 0.4468 * r) ** 2 
    It = 2 / 3 * (b - 0.63 * tf) * tf ** 3 + 1 / 3 * (h - 2 * tf) * tw ** 3 + 2 * (tw / tf) * (0.145 + 0.1 * r / tf) * (((r + tw / 2) ** 2 + (r + tf) ** 2 - r ** 2) / (2 * r + tf)) ** 4
    Iw = tf * b ** 3 / 24 * (h - tf) ** 2
    E=210000 #MPa, because all dimensions are in mm
    C1=1.0
    C2=0.0
    C3=0.0
    kz=1.0
    kw=1.0
    zg=0.0
    zj=0.0
    PI=3.141592653589793
    mcr_u = C1 * PI ** 2 * E * Iz / (kz * L) ** 2 * (((kz / kw) ** 2 * Iw / Iz + (kz * L) ** 2 * E / (2 * (1 + 0.3)) * It / (PI ** 2 * E * Iz) + (C2 * zg - C3 * zj) ** 2) ** 0.5 - (C2 * zg - C3 * zj))
    #output in kNm
    return mcr_u*10**-6
```

```python
# Convert DataFrame to NumPy array
df = pd.read_csv('training_set.csv', sep=";")
df.head()

numpy_array = df.to_numpy()

for j in range(0,len(df)):
    y_norm = df.y[j] / Mcr_u(df.hw1[j], df.hw2[j], df.tw[j],df.bf1[j],df.tf1[j],df.bf2[j],df.tf2[j], df.L[j])
    numpy_array[j][9] = y_norm

# Specify the file path
file_path = 'training_set_normalized_y.csv'

# Save the NumPy array to a CSV file using NumPy's savetxt function
header = ';'.join(df.columns)
np.savetxt(file_path, numpy_array, delimiter=';', header=header, comments='',  fmt='%0.10f')
```
